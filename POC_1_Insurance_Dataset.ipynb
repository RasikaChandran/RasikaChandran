```python
# x,y spilt
x = df.iloc[:,[0,1,2,3,4,5]]
y = df.iloc[:,[6]]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
x['sex'] = le.fit_transform(x['sex'])
x['smoker'] =  le.fit_transform(x['smoker'])
x['region'] = le.fit_transform(x['region'])
x
```

```python
y
```

```python
# standard scalar
from sklearn.preprocessing import StandardScaler
std_d = StandardScaler().fit_transform(x)
```

```python
S = std_d
S
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(S,y,test_size=0.3,random_state = 113)
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
std_reg = LinearRegression()
std_reg.fit(x_train,y_train)
```

```python
y_predict = std_reg.predict(x_test)
```

```python
y_predict
```

```python
corr_df = df[["charges","age","bmi","children"]].corr()
sns.heatmap(corr_df,annot = True , cmap = "coolwarm")
plt.title ("describe analysis")
plt.show()
```

FINDING MAE,MSE,R-SCALE VALUE

```python

from sklearn .metrics import mean_absolute_error
c = std_reg.predict(x_test)
mean_absolute_error(c,y_test)
```

```python
from sklearn.metrics import mean_squared_error
m = std_reg.predict(x_train)
mean_squared_error(m,y_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_predict,y_test)
```
